```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
from pathlib import Path
dir = os.path.realpath('..')
datfile = r'calibration_data/RampTest_2022-01-24_124501.csv'
fname = os.path.join(dir, datfile)
df = pd.read_csv(fname)
```

```python
time = df['Time (s)'][642:-1] - df['Time (s)'][642]
rpm = df['Motor Optical Speed (RPM)'][642:-1]
```

```python
plt.figure(figsize=(4, 3), dpi=180)
plt.plot(time, rpm, 'b-')
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Motor RPM')
plt.grid()
plt.xlim([0,0.5])
plt.ylim([0,12000])

rpm_max = np.max(rpm)
plt.hlines(rpm_max, 0, 0.4, 'r', linestyles='dashed')
plt.hlines(0.632*rpm_max, 0, 0.094, 'r', linestyles='dashed')
plt.vlines(0.094, 0, 0.632*rpm_max, 'r', linestyles='dashed')
```
