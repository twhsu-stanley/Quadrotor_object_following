```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
import os
from pathlib import Path
```

```python
# load IMU data
dir = os.path.realpath('..')
datfile = r'calibration_data/Jzz_test2.csv'
fname = os.path.join(dir, datfile)
df = pd.read_csv(fname)
df['time_s']=(df['imu_time_ns'] - df['imu_time_ns'][0] ) /np.power(10,9)

# parameters
mass = 0.5
g = 9.81
d = 1
L = 1

# period
T = 1 

# Determin the moment of inertia
J = mass * g * d * d / (16 * np.pi * np.pi * L) * T * T
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.plot(df.time_s, df[['roll','pitch','yaw']])
plt.legend(['roll','pitch','yaw'])
plt.ylabel("angle (rad)")
plt.xlabel("time (s)")
plt.grid()
```
